```python
import pandas 

data = pandas.read_csv("datasets/supermarket_data.csv")
data
```

```python
# Use the describe fuction
data.describe()
```

```python
# We have noticed that the column Itemprice, contains 3 empty slots
# We need to fill these empty slots with a value, and in this case we shall fill with the mean of the column
data["itemprice"] = data["itemprice"].fillna(55)

# Describe the data again
data.describe()
```

```python
# Get the number of entries in a given column
data.groupby("gender").size()
```

```python
# Combining multiple columns
data.groupby(["city", "payment"]).size()
```

```python
# Import the modules
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Below is a count plot showing the distribution of males to females
sns.countplot(x="gender", data=data, palette="winter")
plt.title("Gender Distribution")
plt.ylabel("Total count of Individuals")
```

```python
# Create a count plot of product
# Which product category appears the most?
sns.countplot(x = "product", data = data, palette = "Oranges")
plt.title("Product Distribution")
plt.xticks(rotation = 45)
```

```python
# Below is a countplot that has
sns.countplot(x="city", hue="gender", palette="Reds", data=data)
plt.title("Gender Distribution in different cities")
plt.minorticks_on()
plt.grid(True, which='major')
plt.grid(True, which='minor', linestyle=':')
```

```python
# Below is a pieplot
data["payment"].value_counts().plot.pie(autopct='%1.1f%%')
```
